```python
#!pip3 install requests
#!pip install folium
#!pip install plotly
import requests
import json
import pandas as pd
from pandas import json_normalize
import datetime
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#url for Police Department Incident Reports: 2018 to Present (dataset #1)
url = "https://data.sfgov.org/resource/wg3w-h783.json?$limit=300000"
#url for Police Department Incident Reports: Historical 2003 to May 2018 (dataset #2)
url_2 = "https://data.sfgov.org/resource/tmnf-yvry.json?$limit=1200000"
```

```python
#returns a JSON object of the result (url)
response = requests.get(url).json()

#returns a JSON object of the result (url_2)
response_2 = requests.get(url_2).json()
```

```python
#Normalize semi-structured JSON data into a flat table. (df)
res = json_normalize(response)
df = pd.DataFrame(res)

##Normalize semi-structured JSON data into a flat table. (df_2)
res_2 = json_normalize(response_2)
df_2 = pd.DataFrame(res_2)
```

```python
#extract the months from incident date before dropping it (dataset #1)
df['incident_month'] = pd.DatetimeIndex(df['incident_date']).month
```

```python
df = df.drop(['incident_datetime', 'row_id','report_type_description',
'report_datetime','incident_time','incident_date','incident_day_of_week',':@computed_region_nqbw_i6c3',
 ':@computed_region_h4ep_8xdi','report_type_code',
 ':@computed_region_n4xg_c4py',
 ':@computed_region_jg9y_a9du',
 ':@computed_region_jwn9_ihcz',
 ':@computed_region_26cr_cadq','analysis_neighborhood',
 ':@computed_region_qgnn_b9vv', 'police_district','incident_subcategory',
 'point.type', 'supervisor_district', 'resolution','filed_online','incident_number', 'incident_id', 'cad_number', 'cnn', 'intersection','point.coordinates',], axis = 1)
```

```python
#dropping null values
df = df.dropna(how="any")

#dropping 2023 data and reset the index
df = df.drop(df.index[df['incident_year'] == "2023"])
df = df.reset_index(drop = True)

df_2['incident_month'] = pd.DatetimeIndex(df_2['date']).month
df_2['incident_year'] = pd.DatetimeIndex(df_2['date']).year
```

```python
#rename columns (dataset #2)
df_2.rename(columns = {'category':'incident_category', 'descript':'incident_description', 'y':'latitude', 'x':'longitude'}, inplace = True)
#dropping null values (dataset #2)
df_2 = df_2.dropna(how="any")
#changing incident_year type to mearge the two Dataframes (dataset #2)
df_2.astype({'incident_year': 'object'},{}).dtypes
```

```python
#creating a combined Dataframe 
crime_df = pd.concat([df, df_2], ignore_index=True, sort=False)
crime_df.astype({'incident_category': 'str'},{'incident_description' : 'str'}).dtypes
```

```python
# Changing letters to lower case in both incident_category & incident_description columns
crime_df["incident_category"] = crime_df["incident_category"].str.lower()
crime_df["incident_description"] = crime_df["incident_description"].str.lower()

#know all type of unique values in column
crime_df["incident_category"].unique()
```

```python
# replace values.
rep = {
    "drug violation":"drug offense",
    'drug/narcotic':"drug offense",
    "forgery/counterfeiting" : "forgery and counterfeiting",
    "larceny/theft" : "larceny theft",
    "human trafficking, commercial sex acts" : "human trafficking (a), commercial sex acts",
    "motor vehicle theft?" : "motor vehicle theft",
    "other miscellaneous" : "other",
    "other offenses" : "other",
    "sex offenses, forcible" : "sex offense",
    'sex offenses, non forcible' : 'sex offense',
    'suspicious occ' : 'suspicious',
    'warrants': 'warrant',
    'weapons offence': 'weapons offense',
    'weapons carrying etc' : 'weapon laws'
}

crime_df.replace({"incident_category": rep})
```

```python
crime_df['latitude'] = crime_df['latitude'].astype(float)
```

```python
crime_df['longitude'] = crime_df['longitude'].astype(float)
crime_df.dtypes
```

```python
neighborhood_limits=pd.read_csv("SF-neighborhood-limits.csv")
neighborhood_limits
```

```python
Chinatown = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][0]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][0]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][0]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][0])
                    ]
RussianHill = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][1]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][1]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][1]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][1])
                    ]                               

MissionBay = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][2]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][2]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][2]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][2])
                    ]    

NobHill = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][3]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][3]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][3]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][3])
                    ]  
Marina = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][4]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][4]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][4]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][4])
                    ] 

Sunset = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][5]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][5]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][5]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][5])
                    ]
NoeValley = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][6]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][6]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][6]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][6])
                    ]
InnerRichmond = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][7]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][7]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][7]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][7])
                    ]
BernalHeights = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][8]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][8]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][8]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][8])
                    ]
GlennPark = crime_df[(crime_df["latitude"] > neighborhood_limits["Lat-south"][9]) & 
                    (crime_df["latitude"] < neighborhood_limits["Lat-north"][9]) & 
                    (crime_df["longitude"] < neighborhood_limits["Lng-east"][9]) & 
                    (crime_df["longitude"] > neighborhood_limits["Lng-west"][9])
                    ]
```

```python
Chinatown_counts=Chinatown["incident_year"].value_counts()
Chinatown_counts = pd.DataFrame(Chinatown_counts)
Chinatown_counts.rename(columns = {'incident_year':'Chinatown'}, inplace = True)

RussianHill_counts=RussianHill["incident_year"].value_counts()
RussianHill_counts = pd.DataFrame(RussianHill_counts)
RussianHill_counts.rename(columns = {'incident_year':'RussianHill'}, inplace = True)

MissionBay_counts=MissionBay["incident_year"].value_counts()
MissionBay_counts = pd.DataFrame(MissionBay_counts)
MissionBay_counts.rename(columns = {'incident_year':'MissionBay'}, inplace = True)

NobHill_counts=NobHill["incident_year"].value_counts()
NobHill_counts = pd.DataFrame(NobHill_counts)
NobHill_counts.rename(columns = {'incident_year':'NobHill'}, inplace = True)

Marina_counts=Marina["incident_year"].value_counts()
Marina_counts = pd.DataFrame(Marina_counts)
Marina_counts.rename(columns = {'incident_year':'Marina'}, inplace = True)

Sunset_counts=Sunset["incident_year"].value_counts()
Sunset_counts = pd.DataFrame(Sunset_counts)
Sunset_counts.rename(columns = {'incident_year':'Sunset'}, inplace = True)

NoeValley_counts=NoeValley["incident_year"].value_counts()
NoeValley_counts = pd.DataFrame(NoeValley_counts)
NoeValley_counts.rename(columns = {'incident_year':'NoeValley'}, inplace = True)

InnerRichmond_counts=InnerRichmond["incident_year"].value_counts()
InnerRichmond_counts = pd.DataFrame(InnerRichmond_counts)
InnerRichmond_counts.rename(columns = {'incident_year':'InnerRichmond'}, inplace = True)

BernalHeights_counts=BernalHeights["incident_year"].value_counts()
BernalHeights_counts = pd.DataFrame(BernalHeights_counts)
BernalHeights_counts.rename(columns = {'incident_year':'BernalHeights'}, inplace = True)

GlennPark_counts= GlennPark["incident_year"].value_counts()
GlennPark_counts = pd.DataFrame(GlennPark_counts)
GlennPark_counts.rename(columns = {'incident_year':'GlennPark'}, inplace = True)
```

```python
Chinatown_counts.to_csv("Chinatown_counts.csv")
RussianHill_counts.to_csv("RussianHill_counts.csv")
MissionBay_counts.to_csv("MissionBay_counts.csv")
NobHill_counts.to_csv("NobHill_counts.csv")
Marina_counts.to_csv("Marina_counts.csv")
Sunset_counts.to_csv("Sunset_counts.csv")
NoeValley_counts.to_csv("NoeValley_counts.csv")
InnerRichmond_counts.to_csv("InnerRichmond_counts.csv")
BernalHeights_counts.to_csv("BernalHeights_counts.csv")
GlennPark_counts.to_csv("GlennPark_counts.csv")
```
